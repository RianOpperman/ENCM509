## Introduction

The purpose of this lab is to explore and understand the classical approaches to facial feature extraction and matching. These approaches utilize Principal Component Analysis (PCA) for extracting facial features and Euclidean Distance (ED) to calculate a similarity score between feature vectors, as well as thresholds for face matching. Additionally, we will use a simple KNN classifier for matching faces, using the PCA-extracted facial features, to classify input faces.

By using the AT&T Faces Database we can accurately investigate these topics due to the database's high-quality images, which are already pre-processed with a square aspect ratio and are in grayscale colour. 

## Exercise 1

### Creating database

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage import exposure
import scipy.linalg as la

from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix

from utils_cm import plot_confusion_matrix

import matplotlib as mpl
# setting the default colormap for images only
mpl.rc('image', cmap='gray')

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
path = 'ATT dataset'
subjects = [ 7, 11, 23 ]
M = 9
M_total = M * len(subjects)
um = 100
ustd = 80

probe_imgs = [imread(f'{path}/s{subject}/10.pgm', as_gray=True) for subject in subjects ]
irow, icol = probe_imgs[0].shape

# allocation of vector that will have all images
S = np.zeros((irow*icol, M_total)) # img matrix

plt.figure(figsize=(8,8))
for s, subj in enumerate(subjects):
    for i in range(1,M+1):
        img = imread(f'{path}/s{subj}/{i}.pgm', as_gray=True)
        
        plt.subplot(3,3,i)
        plt.imshow(img, cmap='gray')
        plt.axis('off')
        
        # reshape(img',irow*icol,1); 
        # creates a (N1*N2)x1 vector
        temp = np.reshape(img, (irow * icol, 1))
        S[:, s * M + (i - 1)] = temp[:,0] 
    plt.show()
```

```python
# Normalization
for i in range(S.shape[1]):
    temp = S[:,i]
    m = np.mean(temp)
    st = np.std(temp)
    # calculation where are defined the new pixels intensities
    S[:,i] = (temp - m) * ustd / st + um

# Creating covariance matrix
dbx = S.copy()
A = dbx.T
L = np.matmul(A, A.T)

## vv is the eigenvector for L
## dd is the eigenvalue for  L = dbx.T*dbx
dd, vv = la.eig(L)

# Sorting & eliminating zero eigenvalues
v = np.zeros(vv.shape)
d = np.zeros(dd.shape)

for i in range(vv.shape[1]):
    if (dd[i] > 1e-4):
        v[:,i] = vv[:,i]
        d[i] = dd[i]
        
## sort d in ascending order
index = np.argsort(d)
B = d[index]
ind = np.zeros(index.shape).astype(int)
dtemp = np.zeros(index.shape)
vtemp = np.zeros(v.shape)

idx_len = len(index)
for i in range(idx_len):
    dtemp[i] = B[idx_len-1-i]
    ind[i] = idx_len-1-index[i]
    vtemp[:,ind[i]] = v[:,i]

d = dtemp
v = vtemp

# Normalizing eigenvectors
for i in range(v.shape[1]):
    kk = v[:,i]
    temp = np.sqrt(np.sum(np.power(kk,2)))
    v[:,i] = v[:,i] / temp
    
# Eigenvectors of C matrix
u = np.zeros(dbx.shape)
for i in range(v.shape[1]):
    temp = np.sqrt(d[i])
    u[:,i] = np.matmul(dbx, v[:,i]) / temp
    
for i in range(u.shape[1]):
    kk = u[:,i]
    temp = np.sqrt(np.sum(np.power(kk,2)))
    u[:,i] = u[:,i] / temp

# Finding the weight of each face
omega = np.zeros((M_total, M_total))
WW = np.zeros((M_total, 1))

for h in range(dbx.shape[1]):
    for i in range(u.shape[1]):
        t = u[:,i].T
        WeightOfImage = np.dot(t, dbx[:,h].T)
        WW[i] = WeightOfImage
    omega[:,h] = WW[:,0]
```

### Matching face inside database

```python
def match_scores(imgs):
    plt.figure(figsize=(6,6))
    scores = []
    for idx, probe in enumerate(imgs):
        plt.subplot(3, 2, (idx * 2) + 1)
        plt.imshow(probe, cmap='gray')
        plt.title('Input Image')
        plt.axis('off')

        InImage = np.reshape(probe, (irow*icol,1))

        # Normalization
        temp = InImage
        me = np.mean(temp)
        st = np.std(temp)
        temp = (temp-me) * ustd/st + um
        NormImage = temp

        # Finding Difference: NormalizedImage - MeanImage
        Difference = temp - m

        p = np.zeros((u.shape[1],1))
        for i in range(u.shape[1]):
            # dot returns scalar vector product
            pare = np.dot(u[:,i], NormImage) 
            p[i] = pare

        # m is the mean image, u is the eigenvector
        ReshapedImage = m + np.matmul(u, p) 
        ReshapedImage = np.reshape(ReshapedImage, (irow,icol))

        # show the reconstructed image
        plt.subplot(3, 2, (idx * 2) + 2)
        plt.imshow(ReshapedImage, cmap='gray')
        plt.title('Reconstructed Image')
        plt.axis('off')

        InImWeight = np.zeros((M_total,1))

        for i in range(u.shape[1]):
            t = u[:,i].reshape(-1,1).T
            # scalar vector product of vectors t and Difference
            WeightOfInputImage = np.dot(t, Difference) 
            InImWeight[i] = WeightOfInputImage

        e_in_database = np.zeros((M_total, 1))
        for i in range(omega.shape[1]):
            q = omega[:,i].reshape(-1,1)
            DiffWeight = InImWeight - q
            mag = np.linalg.norm(DiffWeight)
            e_in_database[i] = mag

        # max/min Euclidean distance
        scores.append(e_in_database)
        print(f'Probe image {idx + 1}:\n\tMax = {np.max(e_in_database)}\n\tMin = {np.min(e_in_database)}')
    plt.show()
    return scores
```

```python
def plot_scores(ed_scores):
    labels = [ f'{i}' for i in range(1, M_total + 1) ]
    for score in ed_scores:
        plt.bar(np.arange(len(score)), score.flatten(), tick_label=labels)
        # plt.ylim([10000, 16000])
        plt.xlabel('Image #')
        plt.ylabel('Distance')
        plt.title('Euclidean Distances: Same subject\n'\
                'Min {:.2f}, Max {:.2f}'.format(np.min(score), np.max(score)));
        plt.show()

inside_scores = match_scores(probe_imgs)
plot_scores(inside_scores)
```

## Exercise 2

### Experimental threshold 

```python
print(f'Mean of insider distances: {np.mean(inside_scores)}')
print(f'Mean of outsider distances: {np.mean(outside_scores)}')
print(f'Mean of non-face distances: {np.mean(non_scores)}')
```

```python
# Set Threshold Values 
insider_threshold = 12000
outsider_threshold = 12500

def classify(distances, insider_threshold, outsider_threshold):
    mean_of_distances = np.mean(distances);
    if (mean_of_distances <= insider_threshold):
        print('    Image is in the database')
    elif (mean_of_distances > insider_threshold and mean_of_distances <= outsider_threshold):
        print('    Image is a face but not in the database')
    elif (mean_of_distances > outsider_threshold):
        print('    Image is not a face')

print("Inside database:")
classify(inside_scores, insider_threshold, outsider_threshold)
print("Outside database:")
classify(outside_scores, insider_threshold, outsider_threshold)
print("Non-face:")
classify(non_scores, insider_threshold, outsider_threshold)
```

### Automated threshold

```python
def calculate_threshold(scores):
    mean_of_distances = np.mean(scores)
    return round(mean_of_distances * 1.01)

insider_threshold = calculate_threshold(inside_scores)
outsider_threshold = calculate_threshold(outside_scores)

print(f'Calculated inner threshold: {insider_threshold}')
print(f'Calculated outer threshold: {outsider_threshold}\n')

print("Inside database:")
classify(inside_scores, insider_threshold, outsider_threshold)
print("Outside database:")
classify(outside_scores, insider_threshold, outsider_threshold)
print("Non-face:")
classify(non_scores, insider_threshold, outsider_threshold)
```

### Analysis

**Experimental thresholds:**

| Threshold | Mean of distances |
| --- | ---: |
| Insider threshold | 12,000.00 |
| Outsider threshold | 12,500.00 |

**Automated thresholds:**

| Threshold | Mean of distances |
| --- | ---: |
| Insider threshold | 12,023.00 |
| Outsider threshold | 12,361.00 |

As can be seen above, we can determine the thresholds by taking the mean of the computed Euclidean distances, as this allows us to better adapt to a low-quality input image in either the training database or as input to our classification system. Afterwards, we can examine the mean of the thresholds for faces inside & outside the database, as well as the mean of the threshold for non-faces. With this information we can accurately determine an adequate insider & outsider threshold, which will help us to determine if a face is inside or outside the database, or even if the input is not a face at all! 

This has resulted in the thresholds shown in the tables above, i.e. a threshold for inside the database of 12,000 and a threshold for outside the database of 12,500. Using these thresholds, we can determine that if the input image's score is below the insider threshold, the image is inside the database. Also, we know that if the score is between the insider & outsider thresholds the face is not in the database, and if the score is larger than the outside threshold then the image is classified as not-a-face. 

The current major flaw in our implementation is that outliers can really skew the data with our mean-based method, in both training and testing. To improve on this procedure, it would be better to use a more robust algorithm which takes into account all the input data and doesn't allow outliers to skew the data. In this way, it allows the detection system to be quite accurate, even if the input image is a low-quality image of a user inside the database. As well, another improvement would be to automate this threshold calculation, as relying solely on an experimental threshold is tedious and prone to user errors from inaccurate analysis. An example of this process is provided above with our `calculate_threshold()` function, however it always increases the calculated mean by 1% of it's own value to provide a wider margin in case of low-quality images. This lacks the fine-grained fidelity to truly adapt to any database, and as such a more robust procedure would be desired, such as utilizing machine learning to understand the input data more deeply and provide a more apt threshold. 

## Exercise 3

```python
def prediction_evaluation_all_subjects(predicted_label, true_label, num_of_subjects):

    # True Positive (TP): real label IS subject_id and predicted label IS subject_id
    # False Positive (FP): real label is NOT subject_id while predicted label IS subject_id
    # True Negative (TN): real label is NOT subject_id and predicted label is NOT subject_id
    # False Negative (FN): real label IS subject_id and predicted label is NOT subject_id
    
    TP, FP, TN, FN = 0, 0, 0, 0

    for subject_id in range(1,num_of_subjects+1):
        for predictedLabel, realLabel in zip(predicted_label, true_label):
            if predictedLabel == subject_id and realLabel == subject_id:
                TP += 1
            if predictedLabel != subject_id and realLabel == subject_id:
                FN += 1
            if predictedLabel != subject_id and realLabel != subject_id:
                TN += 1
            if predictedLabel == subject_id and realLabel != subject_id:
                FP += 1
    print("Overall:\n\tTP: %d, FP: %d, TN: %d, FN: %d" % (TP, FP, TN, FN))
    
    # True Positive Rate (TPR) also known as Sensitivity or Recall
    TPR = TP / (TP + FN)
    # True Negative Rate (TNR) also known as Specificity
    TNR = TN / (TN + FP)
    # False Positive Rate (FPR)
    FPR = FP / (FP + TN)
    # False Negative Rate (FNR)
    FNR = FN / (TP + FN)

    print("\tTPR: %.2f%%, TNR: %.2f%%, FPR: %.2f%%, FNR: %.2f%%" % (TPR*100, TNR*100, FPR*100, FNR*100))
```

#### increasing number of subjects to allow for PCA with 200 components

```python
# increasing number of subjects to allow for PCA with 200 components
n_subjects = 40

# number of images should sum to 10
n_training_images = 5
n_test_images = 5

knn_neighbors = 3

# You can specify you directory here, please, change the string below:
path = '../ATT dataset/'

print('For EACH of all the %d subjects:' % (n_subjects))
print('\t%d images will be used for TRAINING:' % (n_training_images), [x for x in range(1,n_training_images+1)])
print('\t%d images will be used for TEST:' % (n_test_images), [x for x in range(n_training_images+1, n_training_images+n_test_images+1)])
```

```python
# use the first 5 images from subjectID 1-n_subjects to train the classifier
trainingFaces = []
trainingLabels = []
for subjectId in range(1, n_subjects+1):
    for imageId in range(1, n_training_images+1):
        img = imread("./ATT dataset/s%d/%d.pgm" %(subjectId, imageId), as_gray=True)
        trainingFaces.append(np.reshape(img, (img.size,)))
        trainingLabels.append(subjectId)

# use the last 5 images from subjectID 1-20 to test the classifier
testingFaces = []
testingLabels = []
for subjectId in range(1, n_subjects+1):
    for imageId in range(n_training_images+1, n_training_images+n_test_images+1):
        img = imread("./ATT dataset/s%d/%d.pgm" %(subjectId, imageId), as_gray=True)
        testingFaces.append(np.reshape(img, (img.size,)))
        testingLabels.append(subjectId)

print('Training faces:', len(trainingFaces))
print('Size of each sample:', len(trainingFaces[0]))
```

```python
# n_components: number of principal components
n_components_pca_50 = 50
pca_50 = PCA(n_components=n_components_pca_50)

# fit the model, e.g., creating the covariance matrix... as we manually did in Lab 4
pca_50.fit(trainingFaces)

train_pca_50 = pca_50.transform(trainingFaces)
test_pca_50 = pca_50.transform(testingFaces)

knn_pca_50 = KNeighborsClassifier(n_neighbors=knn_neighbors).fit(train_pca_50, trainingLabels)
predictedLabels_50 = knn_pca_50.predict(test_pca_50)
prediction_evaluation(predictedLabels_50, testingLabels)
prediction_evaluation_all_subjects(predictedLabels_50, testingLabels, n_subjects)
```

```python
# Generate the confusion matrix
confusionMatrix = confusion_matrix(testingLabels, predictedLabels_50)

plot_confusion_matrix(cm=confusionMatrix,
                      target_names = [i for i in range(1, n_subjects+1)])
```
